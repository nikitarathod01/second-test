# Financial Portfolio Analysis

```python
#(1). Load and preprocess the stock price data.

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:\\Edbyte data analysis course\\Financial Portfolio Analysis.csv")
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.head(5)
```

```python
# (2). Calculate daily returns for each stock


df['Daily_Return (%)'] = (df['Purchase Price'] / df['Current Price']) * 100
df
```

```python
df['Weighted_Return'] = df['Daily_Return (%)'] * df['weights']
df
```

```python
# (3). Calculate portfolio returns and risk.
portfolio_returns = df['Weighted_Return'].sum()
portfolio_risk = df['Weighted_Return'].std()

print("Portfolio Returns:", portfolio_returns)
print("Portfolio Risk (Standard Deviation):", portfolio_risk)
```
